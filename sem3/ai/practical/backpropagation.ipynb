```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Input sample and target
X = np.array([[1, 1, 0, 1]])
T = np.array([[1, 0]])

# Build network: 4–3–2–2
model = Sequential([
    Dense(3, activation='sigmoid', input_shape=(4,), name="Hidden_Layer_1"),
    Dense(2, activation='sigmoid', name="Hidden_Layer_2"),
    Dense(2, activation='sigmoid', name="Output_Layer")
])

# Compile model
model.compile(optimizer='sgd', loss='mse')

# Train for 1 epoch
model.fit(X, T, epochs=1, verbose=1)

# Predict output
output = model.predict(X)
print("\nNetwork Output:", output)

# -------------------------------
# Print Weights and Biases
# -------------------------------
print("\n--- Weights and Biases After Training ---")

for layer in model.layers:
    weights, biases = layer.get_weights()

    print(f"\nLayer Name: {layer.name}")
    print("Weights Matrix:")
    print(weights)

    print("Bias Vector:")
    print(biases)
```
