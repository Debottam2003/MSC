```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
```

```python
data = [[1, 2], 
        [2, 3], 
        [3, 4], 
        [8, 8], 
        [9, 10], 
        [10, 12]]

df = pd.DataFrame(data, columns=["X", "Y"])
print(df)
```

```python
agnes = AgglomerativeClustering(n_clusters=1, linkage="single")

labels = agnes.fit_predict(df)

print("Cluster Labels:", labels)
```
